# Convolutional Neural Network

### Importing the libraries

```
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```
tf.__version__
```

## Part 1 - Data Preprocessing

### Preprocessing the Training set

```
#Geometrical tranformations on images of training set: avoid overfitting (image augemntation)
#Look in online Keras API docs
```

```
train_datagen = ImageDataGenerator(
        rescale=1./255, #feature scaling
        shear_range=0.2, #transformations
        zoom_range=0.2,
        horizontal_flip=True)
training_set = train_datagen.flow_from_directory(
        'dataset/training_set', #path to training set
        target_size=(64, 64), #final size of images to fit into network
        batch_size=32, #how many images in the set
        class_mode='binary')
```

### Preprocessing the Test set

```
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        'dataset/test_set', #path to test set
        target_size=(64, 64), #same target size
        batch_size=32,
        class_mode='binary')
```

## Part 2 - Building the CNN

```
#Making the Brain
```

### Initialising the CNN

```
#Create CNN variable
cnn = tf.keras.models.Sequential()
```

### Step 1 - Convolution

```
#add first convoltional layer
cnn.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu', input_shape = [64, 64, 3]))
```

### Step 2 - Pooling

```
#Apply Max Pooling
#Adding pooling layers to convolution layer
```

```
cnn.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
```

### Adding a second convolutional layer

```
cnn.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
```

### Step 3 - Flattening

```
#Make layers into 1D vector
```

```
flatten = cnn.add(tf.keras.layers.Flatten())
```

### Step 4 - Full Connection

```
cnn.add(tf.keras.layers.Dense(units = 128, activation = 'relu'))
```

### Step 5 - Output Layer

```
cnn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

## Part 3 - Training the CNN

```
#Making the brain smart
```

### Compiling the CNN

```
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Training the CNN on the Training set and evaluating it on the Test set

```
cnn.fit(x = training_set, validation_data = test_set, epoch = 25)
```

## Part 4 - Making a single prediction

```
#Predict what's inside the two images of a dog and cat
```

```
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('dataset/single_prediction/cat_or_dog_1.jpg', target_size = [64, 64])
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0) #add extra dimension related to the batch size
result = cnn.predict(test_image) #image now has right format expected by predict method
training_set.class_indices #cat = 0, dog = 1
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'
```
