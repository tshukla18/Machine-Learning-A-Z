# Data Preprocessing Tools

## Importing the libraries

```
#numpy: for arrays
#matplotlib: plots
#pandas: import data sets and create matrix of features (ie. variables);preprocess and import data sets
```

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
#create a new variable that contains the data set with its data frames
#dataset has FEATURES and DEPENDENT VARIABLES
#Features: data that predicts dep. variables
#iloc():locate index (of columns and/or rows)
```

```
dataset = pd.read_csv('Data.csv')
x = dataset.iloc[:, :-1].values #get first three col. and all rows
y = dataset.iloc[:, -1].values #get last column and all rows
```

```
print(x)
```

```
print(y)
```

## Taking care of missing data

```
#replace missing data by average of all the values in that data element
#scikit learn: data science library
#Class: simpleimputer ---> object will allow us to replace blanks in data

#fit(): connects imputer to matrix of features;accepts columns w/ numerical values
#transform():replaces blank w/ the avg.
```

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy = 'mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

```
print(x)
```

## Encoding ccatategorical data

```
# Encode alphabetical, non-numeric values into numeric values
# Encode these values with binary numbers
# X needs to be transformed into an array of features
```

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```
print(x)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
# Train set: train ML models on existing data/observations
# Test set: Evaluate model performance on new observations
# Training observation = 80%, Test observation = 20%
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```
print(x_train)
```

```
print(x_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
# Scaling all features so they all take values in the same scale; ONLY apply to numerical values
# This step AFTER splitting dataset to prevent test set feature leakage (not good since test set is for FUTURE data sets)
# 2 techniques: standardization and normalization (---> normalization works for normal curve data sets)
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
```

```
print(x_train)
```

```
print(x_test)
```
