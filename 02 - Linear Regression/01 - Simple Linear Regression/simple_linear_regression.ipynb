# Simple Linear Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```
# LinearRegression() class from SkLearn module
```

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #get the model
regressor.fit(x_train,y_train) # train the regression model
```

## Predicting the Test set results

```
y_pred = regressor.predict(x_test) #NOT y_test, which contains REAL salaries
```

## Visualising the Training set results

```
# visualize real salaries (red) vs. predicted salaries (blue)
```

```
plt.scatter(x_train, y_train, color = 'red') #real data points
plt.plot(x_train, regressor.predict(x_train), color = 'blue') #predicted linear regression plot
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```
plt.scatter(x_test, y_test, color = 'red') #real data points
plt.plot(x_train, regressor.predict(x_train), color = 'blue') #don't need to change to x_test values b/c will have same regression line 
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
